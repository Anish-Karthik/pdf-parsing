```python
import os
import google.generativeai as generativeai
```

```python
def configure_client(api_key):
    generativeai.configure(
        api_key=api_key
    )
```

```python
api_key = os.getenv("GEMINI_API_KEY")
configure_client(api_key)

model = generativeai.GenerativeModel(
    model_name="gemini-1.5-flash"
)
```

```python
import time
import traceback
def get_response_delayed_prompt(prompt, delay=0.1):
    try:
        time.sleep(delay)
        raw_response = model.generate_content(
            prompt
        )
        return raw_response.text
    except Exception as e:
        print(traceback.format_exc())
        return get_response_delayed_prompt(prompt, delay * 2)
```

```python
import json

quiz_path = "/Users/pranav/GitHub/pdf-parsing/gemini/Neet/53.json"
with open(quiz_path) as f:
    quiz = json.load(f)
```

```python
def get_reading_material(question):
    chat = model.start_chat(history=[])
    global questions
    prompt1 = f"""
    {question["description"]}
    {question["options"][0]["description"]}
    {question["options"][1]["description"]}
    {question["options"][2]["description"]}
    {question["options"][3]["description"]}
    
    help me learn this, keep me engaged so that i understand"""

    prompt2 = """prepare a content based on this so that i will be able to learn it later"""
    chat.send_message(prompt1)
    response_text = chat.send_message(prompt2).text

    print(response_text)
    question["reading_material"] = response_text
    
    questions.append(question)
```

```python
def split_into_batches(array, batch_size=5):
    return [array[i:i + batch_size] for i in range(0, len(array), batch_size)]
```
